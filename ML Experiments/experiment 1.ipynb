```python
# Aim :- Write a function that takes a list of tuples, where each tuple contains a student's name and their score. 
#The function should return a list of tuples sorted in descending order of scores. If two students have the same score, they should be sorted alphabetically by their names.
# Taslim Ansari 221P010
students = [("Alice",88),("Bob", 75),("Charlie",88),("David",93),("Eve", 75), ("Nishad Joglekar", 90)]
print("Input :",students )
# sort students based on score and names
print('Sorted Scores:')
def sort_key(student):
    return (-student[1], student[0])
students.sort(key=sort_key)
print(students)
```

```python
# Aim :- Write a function that takes a dictionary where keys are student names and values are their scores. 
# The function should return list of student names who have highest score.
# If multiple students have highest scores return all names
# Taslim Ansari 221P010
def find_toppers(score_dict):
    if not score_dict:
        return [], None
    max_score = float('-inf')
    toppers_list = []
    for student, score in score_dict.items():
        if score > max_score:
            max_score = score
            toppers_list = [student]
        elif score == max_score:
            toppers_list.append(student)
    return toppers_list, max_score
student_scores = {
    "Alice": 88,
    "Bob": 75,
    "Charlie": 88,
    "Funsuk": 93,
    "Eve": 93,
    "Taslim Ansari 221P010": 93
}
toppers, highest_score = find_toppers(student_scores)
print("Input Scores:", student_scores)
print("Students with the highest score:")
print(f"Highest Score  : {highest_score}")
print(f"Topper(s)      : {', '.join(toppers)}")
```

```python
'''
Write a function that performs the sigmoid activation on a numpy array 
or scalar representing a layer's output in a neural network. 
The sigmoid function is defined as:
sigmoid(x) = 1 / (1 + e^(-x))
The function should handle both scalar and numpy array inputs.
'''
# Taslim Ansari 221P010
print('By Taslim Ansari 221P010')
def sigmoid(x):
    """Applies the sigmoid activation function to a scalar or numpy array."""
    return 1 / (1 + np.exp(-x))
scalar_value = 0.6
array_input = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 0.6 , 0.7])
print(f'\nScalar input  = {scalar_value}')
print(f'Scalar output = {sigmoid(scalar_value)}')
print(f'\nArray input   = {array_input}')
print(f'Array output  = {sigmoid(array_input)}')
```

```python
'''
Aim: You have a dataset of customer information and their purchasing behavior.
The dataset includes columns such as customer_id, age, gender, annual_income, purchase_amount, and purchase_date.
You need to preprocess this data to prepare it for machine learning tasks:
- Handle missing values
- Encode categorical variables
- Normalize numerical variables
- Create a new feature: the total purchase amount for each customer
'''
# Taslim Ansari 221P010
from sklearn.preprocessing import StandardScaler
# Sample dataset
data = {
    'customer_id': [1, 2, 3, 4, 5, 6],
    'age': [25, 45, np.nan, 35, 50, 22],
    'gender': ['Male', 'Female', 'Female', np.nan, 'Male', 'Male'],
    'annual_income': [50000, 60000, 45000, 80000, 120000, 100000],
    'purchase_amount': [200, 150, 300, 400, np.nan, 1000],
    'purchase_date': ['2023-01-01', '2023-02-15', '2023-01-20', '2023-03-05', '2023-02-25', '2024-07-23']
}
# Create DataFrame
df = pd.DataFrame(data)
# 1. Handle missing values
df['age'].fillna(df['age'].mean(), inplace=True)
df['purchase_amount'].fillna(df['purchase_amount'].mean(), inplace=True)
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
# 2. Encode categorical variable 'gender'
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
# 3. Normalize numerical columns
scaler = StandardScaler()
df[['age', 'annual_income', 'purchase_amount']] = scaler.fit_transform(
    df[['age', 'annual_income', 'purchase_amount']]
)
# 4. Create new feature: total purchase per customer
# (In this mock dataset, each customer appears only once â€” we simulate multiple purchases using groupby)
df['purchase_date'] = pd.to_datetime(df['purchase_date'])  # Convert to datetime
df_total = df.groupby('customer_id', as_index=False)['purchase_amount'].sum()
df_total.rename(columns={'purchase_amount': 'total_purchase'}, inplace=True)
# Merge back into original DataFrame
df = pd.merge(df, df_total, on='customer_id', how='left')
# Final output
print('By Taslim Ansari 221P010\n')
df
```

```python
'''
Aim: You are given a dataset containing information about a machine learning model's performance over multiple epochs
The dataset includes the loss and accuracy of the model for both the training and validation sets.
Your task is to visualize this data using Matplotlib to better understand how the model's performance evolves over ti
'''
# Taslim Ansari 221P010
data = {
'epoch': np.arange(1, 21),
'train_loss': np.random.uniform(0.2, 0.6, 20),
'val_loss': np.random.uniform(0.3, 0.7, 20),
'train_accuracy': np.random.uniform(0.7, 0.95, 20),
'val_accuracy': np.random.uniform(0.6, 0.9, 20)
}
df = pd.DataFrame(data)
df = df.sort_values('epoch')
plt.figure(figsize = (12, 5))
plt.subplot(1, 2, 1)
plt.plot(df['epoch'],df['train_loss'] ,label = "Train Loss", color = "pink", marker ="o" )
plt.plot(df['epoch'],df['val_loss'] ,label = "Value Loss", color = "blue", marker ="o" )
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Vizulizations By Taslim Ansari 221P010')
plt.legend()
# Plot 2
plt.subplot(1, 2, 2)
plt.plot(df['epoch'],df['train_accuracy'] ,label = "Train accuracy", color = "green", marker ="o" )
plt.plot(df['epoch'],df['val_accuracy'] ,label = "Value accuracy", color = "red", marker ="o" )
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training accuracy Vizulizations By Taslim Ansari 221P010')
plt.legend()
```
